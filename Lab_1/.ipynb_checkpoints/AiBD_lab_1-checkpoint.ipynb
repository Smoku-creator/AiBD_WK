Zadanie nr 3:

```python
#Definiowanie funkcji analitycznej
parabola= lambda x: x*x + 5

#Definiowanie funkcji rysującej
def draw(coord: list[int], func):
    #Stworzenie wektora argumentów równoodległych w przedziale coord
    vector = np.linspace(coord[0], coord[1])
    #Inicjalizacja obszaru rysowania
    plt.figure()
    #Tworzenie wykresu z wykorzystaniem argumentu funkcyjnego
    plt.plot(vector, func(vector), 'k-')
    #Generowanie legendy z wykorzystaniem matematycznych symboli pakietu matplotlib.pyplot
    plt.legend([r'$y = x^2 + 5$'])
    #Generowanie tytułu wykresu wykorzystującego formatowanie tekstu
    plt.title(f'Przebieg funkcji w przedziale [{coord[0]}, {coord[1]}]')
    #Dodanie opisów obu osi
    plt.xlabel('Argument x')
    plt.ylabel(r'Wartość funkcji $y = f(x)$')
    #Zastosowanie siatki
    plt.grid()
    #Wyrysowanie wykresu
    plt.show()
```

```python
list_of_coords = [[-1,1], [-6,6], [0,5]]
for coord_i in list_of_coords:
    draw(coord_i, parabola)
```
